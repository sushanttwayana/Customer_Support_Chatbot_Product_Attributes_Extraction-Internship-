# Preprocess the data

```python
import re 
import os 

    


# function to identify invalid product names 
def is_invalid_product(product):
    
    """Check if the product name is invalid(null, empty, too short or ambigous)
    
    Returns : bool: True if invalid, False if valid
    """
    
    # check for null, empty or whitespace
    if pd.isna(product) or str(product).strip() == "":
        return True
    
    product = str(product).strip()
    
    
    # check for single small letter words 
    words = product.split()
    
    if len(words)<=1 or len(product) <= 5:
        return True
    
    # Check for unexpected patterns (e.g., only numbers, random strings)
    if re.match(r'^\d+$', product) or re.match(r'^[a-zA-Z0-9]{1,3}$', product):
        return True
    
    return False
    

    

def preprocess_dataframe(df_main):
    
    # preprocess the dataframe by removing invalid product names returns tuple(cleaned df, removed dataframe)
    
    
    df = df_main.copy()
    
    # flag invalid products
    df['is_invalid'] = df['Products'].apply(is_invalid_product)
    

    # split into valid and invalid dataframes
    df_invalid = df[df['is_invalid']].copy()
    df_valid = df[~df['is_invalid']].copy()
    
    # log invalid entries to a file
    if not df_invalid.empty:
        invalid_file = "invalid_products.csv"
        df_invalid.to_csv(invalid_file, index=False)
        print(f"Invalid proudcts saved to {invalid_file}")
        print(f"Number of invalid producs : {len(df_invalid)}")
        
    return df_valid, df_invalid
    
```

```python
# Appply preprocessing

print(f"original dataframe size:{len(df_main)}")

df_valid, df_invalid = preprocess_dataframe(df_main)

print(f"original dataframe size:{len(df_valid)}")
```

```python
import pandas as pd

def get_middle_rows(df, n=50):
    """
    Returns the middle n rows of a DataFrame.

    Args:
        df: The input DataFrame.
        n: The number of middle rows to return. Default is 5.

    Returns:
        A DataFrame containing the middle rows.
    """
    total_rows = len(df)
    start_index = (total_rows - n) // 2
    end_index = start_index + n

    return df.iloc[start_index:end_index]


middle_df = get_middle_rows(df_valid)
middle_df["Products"].head(50)
```

# More preprocessing

```python
# function to determine if product is vaping-related

def is_vaping_related(product:str) -> bool:
    
    """
    Checks if a product is realted to vvaping, CDB, Kratom or similar categories
    """
    
    product = normalize_product_name(product)
    
    vaping_keywords = [
        "DISPOSABLE", "ELIQUID", "E LIQUID", "E-LIQUID", "E JUICE", "E-JUICE", "COILS", 
        "BATTERY", "VAPE", "VAPES", "VAPORIZER", "POD", "MOD", "KIT", "TANK", "NICOTINE",
        "CBD", "KRATOM", "THCA", "THC", "PRE ROLL", "PREROLL", "FLOWER", "PUFFS", 
        "HOOKAH", "SHISHA", "CIGAR", "CIGARILLO", "PUFFCO", "WATER PIPE", "DAB RIG", 
        "BANGER", "BOWL", "GLASS ON GLASS", "SUB OHM", "POD SYSTEM"
    ]
    non_vaping_keywords = [
        "GUMMIES", "SNACK", "CANDY", "CHEESE", "SHAMPOO", "POUCHES", "PET CARRIER", 
        "BACKPACK", "CLOTHING", "T-SHIRT", "HOODIE", "SWEATS", "CREWNECK", "SPRAY", 
        "AROMA", "AIR FRESHENER", "GRINDER", "JAR", "CLEANER", "DETOX", "COFFEE", 
        "JUICE", "SODA", "CHAMOY", "TABLETS", "PILLS", "MUSHROOM"
    ]
    
    is_vaping = any(keyword in product for keyword in vaping_keywords)
    is_non_vaping = any(keyword in product for keyword in non_vaping_keywords)
                    
                        
    # when both vaping and non vaping keywords are present 
    if is_vaping and is_non_vaping:
        if any(kw in product for kw in ["VAPE", "DISPOSABLE", "ELIQUID", "THCA", "CBD", "KRATOM", "HOOKAH"]):
            return True
        return False
    
    return is_vaping

                        
    
    
    
# function to normalize product names

def normalize_product_name(product:str) -> str:
    """
    Normalize product name by removing extra spaces, special characters, and common prefixes/suffixes.
    """
    
    if not isinstance(product, str):
        product = str(product)
        
    product = product.strip().upper()
    product = re.sub(r'\s+', ' ', product)  # Replace multiple spaces with single space
    product = product.strip('"\'')
    
    # Remove common prefixes/suffixes
    prefixes_suffixes = [
        "NEW ARRIVALS", "SALE", "BEST SELLERS", "EXCLUSIVE", "BOX OF \d+", "PACK OF \d+",
        "\d+ PACK", "SQUARE DISTRIBUTION", "WHOLESALE",
        "BY [A-Z]+", "BOX", "CASE", "DISPLAY", "AUTHENTIC PRODUCTS", "EXPRESS SHIPPING"
    ]
    for pattern in prefixes_suffixes:
        product = re.sub(rf'^{pattern}\s*|\s*{pattern}$', '', product, flags=re.IGNORECASE)
    
    return product.strip()
                        

# Function to extract category key for grouping
def get_category_key(product: str) -> str:
    """
    Extract a category key for grouping similar products based on brand and product type.
    
    Args:
        product: Product name (str).
    
    Returns:
        str: Category key (e.g., 'Geek Bar-DISPOSABLE').
    """
    product = normalize_product_name(product)
    
    # Extract brand (first 1-3 words, avoiding keywords)
    words = product.split()
    brand = words[0] if words else "UNKNOWN"
    if len(words) > 1 and not re.match(r'^(DISPOSABLE|ELIQUID|COILS|BATTERY|PRE[ -]?ROLL|FLOWER|PUFFS|%|HOOKAH|SHISHA|GUMMIES|KRATOM|CIGAR|WATER PIPE|DAB RIG|[0-9])', words[1], re.IGNORECASE):
        brand += " " + words[1]
        if len(words) > 2 and not re.match(r'^(DISPOSABLE|ELIQUID|COILS|BATTERY|PRE[ -]?ROLL|FLOWER|PUFFS|%|HOOKAH|SHISHA|GUMMIES|KRATOM|CIGAR|WATER PIPE|DAB RIG|[0-9])', words[2], re.IGNORECASE):
            brand += " " + words[2]
    
    # Extract product type
    product_type = "OTHER"
    if "DISPOSABLE" in product or "PUFFS" in product:
        product_type = "DISPOSABLE"
    elif any(kw in product for kw in ["ELIQUID", "E LIQUID", "E-LIQUID", "E JUICE", "E-JUICE"]):
        product_type = "ELIQUID"
    elif "COILS" in product:
        product_type = "COILS"
    elif "BATTERY" in product or "VAPORIZER" in product:
        product_type = "BATTERY"
    elif "PRE ROLL" in product or "PREROLL" in product:
        product_type = "PRE ROLL"
    elif "FLOWER" in product:
        product_type = "FLOWER"
    elif "HOOKAH" in product or "SHISHA" in product:
        product_type = "HOOKAH"
    elif any(kw in product for kw in ["CIGAR", "CIGARILLO", "BACKWOODS", "FRONTO", "DUTCH MASTERS"]):
        product_type = "CIGAR"
    elif "KRATOM" in product or any(kw in product for kw in ["MIT 45", "ZANA", "HYDROXIE", "7-OHMZ"]):
        product_type = "KRATOM"
    elif any(kw in product for kw in ["WATER PIPE", "DAB RIG", "BANGER", "BOWL", "GLASS ON GLASS"]):
        product_type = "WATER PIPE"
    elif "POD" in product and "SYSTEM" in product:
        product_type = "POD SYSTEM"
    elif "MOD" in product or "KIT" in product:
        product_type = "MOD KIT"
    
    return f"{brand}-{product_type}"
                        

                        

                        
    
```

```python
from typing import List, Tuple
## function to preprocess, categorie and filter proiducts

def preprocess_and_categorize(df: pd.DataFrame) -> Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    """
    Preprocess DataFrame: deduplicate based on Products and website, categorize, filter non-vaping items,
    and create a final combined DataFrame with vaping_status flag.
    
    Args:
        df: Input DataFrame with 'Products' and 'website' columns.
    
    Returns:
        Tuple: (vaping-related categorized DataFrame, non-vaping DataFrame, final combined DataFrame)
    """
    df = df.copy()
    
    # Ensure 'website' column exists
    if 'website' not in df.columns:
        raise ValueError("DataFrame must contain a 'website' column")
    
    # Normalize product names
    df['Products'] = df['Products'].apply(normalize_product_name)
    
    # Remove duplicates based on both Products and website
    print(f"Original DataFrame size: {len(df)}")
    df_dedup = df.drop_duplicates(subset=['Products', 'website']).copy()
    print(f"After removing duplicates (Products and website): {len(df_dedup)}")
    
    # Add vaping-related flag
    df_dedup['is_vaping_related'] = df_dedup['Products'].apply(is_vaping_related)
    
    # Split into vaping and non-vaping DataFrames
    df_vaping = df_dedup[df_dedup['is_vaping_related']].copy()
    df_non_vaping = df_dedup[~df_dedup['is_vaping_related']].copy()
    
    # Save non-vaping products for review
    if not df_non_vaping.empty:
        non_vaping_file = "non_vaping_products.csv"
        df_non_vaping.to_csv(non_vaping_file, index=False)
        print(f"Non-vaping products saved to {non_vaping_file}")
        print(f"Number of non-vaping products: {len(df_non_vaping)}")
    
    # Add category key and name length to vaping DataFrame
    df_vaping['category_key'] = df_vaping['Products'].apply(get_category_key)
    df_vaping['name_length'] = df_vaping['Products'].str.len()
    
    # Group by category_key and select the longest product name
    df_categorized = df_vaping.sort_values('name_length', ascending=False).groupby('category_key').first().reset_index()
    
    # Save categorized DataFrame for LLM processing
    categorized_file = "categorized_products.csv"
    df_categorized.to_csv(categorized_file, index=False)
    print(f"Vaping-related categorized DataFrame saved to {categorized_file}")
    
    # Log duplicates for review (based on Products and website)
    duplicate_file = "duplicate_products.csv"
    df_duplicates = df[df.duplicated(subset=['Products', 'website'], keep=False)].copy()
    if not df_duplicates.empty:
        df_duplicates.to_csv(duplicate_file, index=False)
        print(f"Duplicate products (same Products and website) saved to {duplicate_file}")
    
    # Save mapping of original products to categories
    df_valid = df.copy()
    df_valid['category_key'] = df_valid['Products'].apply(get_category_key)
    category_mapping_file = "category_mapping.csv"
    df_valid[['Products', 'website', 'category_key']].to_csv(category_mapping_file, index=False)
    print(f"Category mapping saved to {category_mapping_file}")
    
    # Combine categorized and non-vaping DataFrames with vaping_status flag
    df_categorized['vaping_status'] = 'Vaping'
    df_non_vaping['vaping_status'] = 'Non-Vaping'
    
    # Ensure consistent columns (drop name_length from df_categorized to align with df_non_vaping)
    common_columns = list(set(df_categorized.columns).intersection(set(df_non_vaping.columns)))
    df_final = pd.concat([
        df_categorized[common_columns],
        df_non_vaping[common_columns]
    ], ignore_index=True)
    
    # Save final combined DataFrame
    final_file = "final_preprocessed.csv"
    df_final.to_csv(final_file, index=False)
    print(f"Final combined DataFrame saved to {final_file}")
    print(f"Final DataFrame size: {len(df_final)}")
    
    return df_categorized, df_non_vaping, df_final
```

```python
# Apply preprocessing and categorization
df_categorized, df_non_vaping, df_final = preprocess_and_categorize(df_valid)
print(f"Vaping-related categorized DataFrame size: {len(df_categorized)}")
print(f"Non-vaping DataFrame size: {len(df_non_vaping)}")
print(f"Final combined DataFrame size: {len(df_final)}")

# Display sample of final combined DataFrame
print("\nSample of final combined DataFrame:")
print(df_final[['Products', 'website', 'vaping_status', 'category_key']].head())

# Display sample of categorized DataFrame
print("\nSample of vaping-related categorized DataFrame:")
print(df_categorized[['Products', 'website', 'category_key', 'name_length']].head())

# Display sample of non-vaping DataFrame
if not df_non_vaping.empty:
    print("\nSample of non-vaping products:")
    print(df_non_vaping[['Products', 'website', 'is_vaping_related']].head())
```
